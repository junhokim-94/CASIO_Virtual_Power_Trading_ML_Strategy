```python
def rtm_lag_with_auction_logic(group, lags=[24, 48, 72, 96, 120, 144, 168]):
    for lag in lags:
        shifted = group['RTM_mean'].shift(lag)
        shifted_t2 = group['RTM_mean'].shift(lag + 24)  # T-2 is lag + 24 hours

        # If the timestamp is at or after 9am, use T-2 instead of T-1
        use_t2 = group['Time'].dt.hour >= 9
        final = shifted.copy()
        final[use_t2] = shifted_t2[use_t2]

        group[f'RTM_mean_lag_{lag}'] = final
    return group

# Apply the function per location
df = df.groupby('Location').apply(rtm_lag_with_auction_logic).reset_index(drop=True)
```

```python
df_with_lags['Time'] = pd.to_datetime(df_with_lags['Time'])
def rtm_window_same_hour_logic_faster(df, target_cols, rolling_windows=[3, 6]):
    df = df.sort_values('Time').copy()
    df['Hour'] = df['Time'].dt.hour

    for col in target_cols:
        for window in rolling_windows:
            # Rolling mean by location
            df[f'{col}_rolling_mean'] = (
                df
                .groupby(['Location'])[col]
                .transform(lambda x: x.rolling(window=window).mean())
            )

            # Shift by 24h and 48h
            df[f'{col}_roll{window}h_T1'] = df[f'{col}_rolling_mean'].shift(24)
            df[f'{col}_roll{window}h_T2'] = df[f'{col}_rolling_mean'].shift(48)

            # Select T1 or T2 based on auction time logic
            use_t2 = df['Hour'] >= 9
            df[f'{col}_samehour_roll{window}h'] = df[f'{col}_roll{window}h_T1']
            df.loc[use_t2, f'{col}_samehour_roll{window}h'] = df.loc[use_t2, f'{col}_roll{window}h_T2']

            # Drop intermediate columns
            df.drop(columns=[f'{col}_rolling_mean', f'{col}_roll{window}h_T1', f'{col}_roll{window}h_T2'], inplace=True)

    df.drop(columns='Hour', inplace=True)
    return df




def add_prev_day_8am_feature(group, target_cols):
    group['Time'] = pd.to_datetime(group['Time'])
    group['Date'] = group['Time'].dt.date
    group['Hour'] = group['Time'].dt.hour

    for col in target_cols:
        # Extract 8AM values and associate with entire next day
        eight_am_vals = group[group['Hour'] == 8][['Date', 'Location', col]].copy()
        eight_am_vals['Date'] = pd.to_datetime(eight_am_vals['Date']) + pd.Timedelta(days=1)
        eight_am_vals = eight_am_vals.rename(columns={col: f'{col}_prev_8am', 'Date': 'Effective_Date'})

        # Create a column for current rows to join on
        group['Effective_Date'] = pd.to_datetime(group['Date'])

        # Merge on (Effective_Date, Location) to apply 8AM value to all rows of the next day
        group = pd.merge(group, eight_am_vals, on=['Effective_Date', 'Location'], how='left')

    return group.drop(columns=['Date', 'Hour', 'Effective_Date'])



target_cols = [
    'RTM_Load', 'RTM_Solar', 'RTM_Wind', 'RTM_Geothermal', 'RTM_Biomass',
    'RTM_Biogas', 'RTM_Small Hydro', 'RTM_Coal', 'RTM_Nuclear',
    'RTM_Natural Gas', 'RTM_Large Hydro', 'RTM_Batteries', 'RTM_Imports', 'RTM_Other'
]

df_with_lags = df_with_lags.groupby('Location').apply(lambda g: add_prev_day_8am_feature(g, target_cols)).reset_index(drop=True)
df_with_lags = df_with_lags.groupby('Location').apply(lambda g: rtm_window_same_hour_logic_faster(g, target_cols)).reset_index(drop=True)
df_with_lags.tail(50)
```

```python
rtm_real_time_cols = [
    'RTM_Load', 'RTM_Solar', 'RTM_Wind', 'RTM_Geothermal', 'RTM_Biomass',
    'RTM_Biogas', 'RTM_Small Hydro', 'RTM_Coal', 'RTM_Nuclear',
    'RTM_Natural Gas', 'RTM_Large Hydro', 'RTM_Batteries', 'RTM_Imports', 'RTM_Other'
]

# Drop these columns
df_with_lags = df_with_lags.drop(columns=rtm_real_time_cols)
df_with_lags.head(50)
```

```python
import pandas as pd

# Ensure time columns exist
df_with_lags['Time'] = pd.to_datetime(df_with_lags['Time'])
df_with_lags['hour'] = df_with_lags['Time'].dt.hour
df_with_lags['dayofweek'] = df_with_lags['Time'].dt.dayofweek

# Create dummies individually
hour_dummies = pd.get_dummies(df_with_lags['hour'], prefix='hour')
day_dummies = pd.get_dummies(df_with_lags['dayofweek'], prefix='day')
loc_dummies = pd.get_dummies(df_with_lags['Location'], prefix='loc')

# Drop original columns
df_with_lags = df_with_lags.drop(columns=['hour', 'dayofweek', 'Location'])

# Combine into final df
df_final = pd.concat([df_with_lags, hour_dummies, day_dummies, loc_dummies], axis=1)

# Drop rows with missing features (but not targets)
feature_cols = df_final.columns.difference(['DAM_LMP', 'RTM_mean'])
df_final = df_final.dropna(subset=feature_cols)

# Final X, y splits
X_dam = df_final.drop(columns=['DAM_LMP', 'RTM_mean'])
y_dam = df_final['DAM_LMP']

X_rtm = df_final.drop(columns=['DAM_LMP', 'RTM_mean'])
y_rtm = df_final['RTM_mean']
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

def train_and_evaluate(model, X_train, y_train, X_test, y_test, name="Model"):
    print(f"\n🔹 {name} Performance")
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mse = mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    print(f"  MSE: {mse:.4f}")
    print(f"  R² Score: {r2:.4f}")

# Define and train XGBoost for DAM
xgb_dam = xgb.XGBRegressor(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
train_and_evaluate(xgb_dam, X_train_dam, y_train_dam, X_test_dam, y_test_dam, name="XGBoost - DAM")

# Define and train XGBoost for RTM
xgb_rtm = xgb.XGBRegressor(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
train_and_evaluate(xgb_rtm, X_train_rtm, y_train_rtm, X_test_rtm, y_test_rtm, name="XGBoost - RTM")
```

```python
import numpy as np

# Initialize profit to 0
forecast_df['profit'] = 0.0

# Define long condition
long_cond = (
    (forecast_df['Forecast_DAM_LMP'] >= forecast_df['Actual_DAM_LMP']) &
    (forecast_df['Forecast_RTM_mean'] > forecast_df['Forecast_DAM_LMP'] + 1)
)

# Define short condition
short_cond = (
    (forecast_df['Forecast_DAM_LMP'] < forecast_df['Actual_DAM_LMP']) &
    (forecast_df['Forecast_RTM_mean'] < forecast_df['Forecast_DAM_LMP'] - 1)
)

# Calculate profits
forecast_df.loc[long_cond, 'profit'] = (
    forecast_df.loc[long_cond, 'Actual_RTM_mean'] - forecast_df.loc[long_cond, 'Actual_DAM_LMP']
)

forecast_df.loc[short_cond, 'profit'] = (
    forecast_df.loc[short_cond, 'Actual_DAM_LMP'] - forecast_df.loc[short_cond, 'Actual_RTM_mean']
)

# Add Boolean flags for position type
forecast_df['is_long'] = long_cond
forecast_df['is_short'] = short_cond

# Add single position label
forecast_df['position'] = np.where(long_cond, 'long', np.where(short_cond, 'short', 'none'))

forecast_df.head()
```
